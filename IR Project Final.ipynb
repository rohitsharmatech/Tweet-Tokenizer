```python
#We used lists, bad, metions, remove1 and remove2 and appended each with words that don't occur in english dictionary
#bad for words starting with \x - written in different language but it shows as a combination of non meaningful chars
#mentions for any word that starts with f0, for the same reason
#remove1 and remove2 for 'lmao' and for links

bad = []
mentions = []

#We load the csv file with all the tweets in a data frame df
df = pd.read_csv('ua.csv', names=['time', 'tweets', 'topic'])

#We drop all the rows with at least one null value
df = df.dropna(axis=0, how ='any')

#We give each column a header and display the first 5 rows of our data frame df
df.head()
```

In the next section, we append each list with the words supposed to not be in the tweets after cleaning. Because those words don't add any meaning to our summary.

In this section of script, we are trying to remove all the special characters that add no meaning to the summary. Hence, it is really important to remove them.

```python
#To remove all the symbols that are beyond 128 in the ascii sheet
def removetext(text):
    return ''.join([i if ord(i) < 128 else '' for i in text])

#Here we actually remove them
df['tweets'] = df['tweets'].apply(removetext)


#Removing the \n 
df['tweets'] = df['tweets'].apply(lambda x: x.replace('\\n',' '))

#All of these are sef explanatory, we are replacing special symbols by no space.
df['tweets'] = df['tweets'].apply(lambda x: x.replace('!',''))
df['tweets'] = df['tweets'].apply(lambda x: x.replace('"',''))
df['tweets'] = df['tweets'].apply(lambda x: x.replace(';',''))
df['tweets'] = df['tweets'].apply(lambda x: x.replace(':',''))
df['tweets'] = df['tweets'].apply(lambda x: x.replace('&amp',''))
df['tweets'] = df['tweets'].apply(lambda x: x.replace('/',''))
df['tweets'] = df['tweets'].apply(lambda x: x.replace("b'" ,''))
df['tweets'] = df['tweets'].apply(lambda x: x.replace('rt',''))
df['tweets'] = df['tweets'].apply(lambda x: x.replace('`',''))
df['tweets'] = df['tweets'].apply(lambda x: x.replace('-',', '))
```

```python
#All the words that were stored in the lists we formed above will be removed here
for line in df['tweets']:
    for word in line.split():
        if word in bad:
            df['tweets'] = df['tweets'].apply(lambda x: x.replace(word,' '))
        elif word in mentions:
            df['tweets'] = df['tweets'].apply(lambda x: x.replace(word,' '))
```

We use a scoring function to score each sentence and its importance based on the frequency of word that not found in the stop words of english dictionary. Then, we calculate weighted frequency of each of the sentences and the sentences with the most score and length less than that of 40 words each gets printed. This is done to make sure the sentences are not too long.

```python
import nltk

#Scoring a function that is used to convert our paragraph string into a summary
def scoring(paragraph):
    
    #At this point we have preprocessed the data. Next, we need to tokenize the article into sentences. 
    sentence_list = nltk.sent_tokenize(paragraph)
    
    #Now, we are going to find the weighted frequency of occurence
    
    #We first store all the English stop words from the nltk library into a stopwords variable. 
    stopwords = nltk.corpus.stopwords.words('english')
    

    word_frequencies = {}
    
    #To find the frequency of occurrence of each word, we use the paragraph variable. 
    #We used this variable to find the frequency of occurrence since it doesn't contain 
    #punctuation, digits, or other special characters. 
    for word in nltk.word_tokenize(paragraph):
        if word not in stopwords:
            #If not, we proceed to check whether the words exist in word_frequency dictionary 
            #i.e. word_frequencies, or not. 
            if word not in word_frequencies.keys():
                
                #If the word is encountered for the first time, it is added to the dictionary 
                #as a key and its value is set to 1, otherwise updated by 1.
                word_frequencies[word] = 1
            else:
                word_frequencies[word] += 1
                
                
    #Finally, to find the weighted frequency, we can simply divide the number of occurances of 
    #all the words by the frequency of the most occurring word,
    maximum_frequncy = max(word_frequencies.values())
    for word in word_frequencies.keys():
        word_frequencies[word] = (word_frequencies[word]/maximum_frequncy)
    
    
    #Calculating sentence scores
    
    #we first create an empty sentence_scores dictionary. The keys of this dictionary will be 
    #the sentences themselves and the values will be the corresponding scores of the sentences. 
    sentence_scores = {}
    
    #we loop through each sentence in the sentence_list and tokenize the sentence into words.
    for sent in sentence_list:
        for word in nltk.word_tokenize(sent.lower()):
            
            #We then check if the word exists in the word_frequencies dictionary.
            if word in word_frequencies.keys():
                
                #We do not want very long sentences in the summary, therefore, 
                #we calculate the score for only sentences with less than 40 words
                if len(sent.split(' ')) < 40:
                    if sent not in sentence_scores.keys():
                        sentence_scores[sent] = word_frequencies[word]
                    else:
                        sentence_scores[sent] += word_frequencies[word]
    
    #To summarize the article, we can take top N sentences with the highest scores. 
    #The following script retrieves top 7 sentences and prints them on the screen.
    import heapq
    summary_sentences = heapq.nlargest(7, sentence_scores, key=sentence_scores.get)

    summary = ' '.join(summary_sentences)
    print(summary)
```
